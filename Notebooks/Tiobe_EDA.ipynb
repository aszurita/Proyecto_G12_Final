```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
# Cargar datos
df = pd.read_csv('../Datos/Series_de_Tiempo.csv')
df['Date'] = pd.to_datetime(df['Date'])
# Añadir columna de año 
df['Year'] = df['Date'].dt.year
df = df[(df['Year'] >= 2020) & (df['Year'] <= 2025)]
df.reset_index().to_csv('../Datos/Series_de_Tiempo.csv', index=False)
# Lenguajes de interés
languages = [
    "Python",
    "C",
    "C#",
    "C++",
    "Java",
    "JavaScript",
    "Assembly",
    "R",
    "Perl",
    "Fortran",
    "Rust",
    "MATLAB",
    "PHP",
    "Go",
    "Kotlin",
]
df = df[df['Language'].isin(languages)]
# Crear tabla pivote
pivot_table = df.pivot_table(
    values='Rating',
    index='Language',
    columns='Year',
    aggfunc='mean'
).round(2)

# Guardar
# pivot_table.to_csv('rating_promedio.csv')
pivot_table.reset_index().to_csv('../Datos_procesados/Rating_promedio.csv', index=False)
```

```python
from dash import dash_table
import pandas as pd

def getIndicadorAnio(df, anio1="2024", anio2="2025"):
    df_anio = df[['Language', anio1, anio2]].copy()
    df_anio['Indicador'] = (df_anio[anio2] - df_anio[anio1]).round(2)

    table = dash_table.DataTable(
        columns=[
            {"name": "Lenguaje", "id": "Language"},
            {"name": anio1, "id": anio1, "type": "numeric"},
            {"name": anio2, "id": anio2, "type": "numeric"},
            {"name": "Indicador", "id": "Indicador", "type": "numeric"},
        ],
        data=df_anio.to_dict("records"),
        sort_action="native",
        sort_mode="single",
        style_table={"overflowX": "auto"},
        style_cell={
            "padding": "10px",
            "fontFamily": "Arial",
            "fontSize": "14px",
            "textAlign": "left",
        },
        style_header={
            "backgroundColor": "#E0F2F1",
            "fontWeight": "bold",
            "border": "none",
        },
        style_data_conditional=[
            {
                "if": {
                    "filter_query": "{Indicador} > 0",
                    "column_id": "Indicador",
                },
                "backgroundColor": "#E8F5E9",
                "color": "#1B5E20",
                "fontWeight": "bold",
            },
            {
                "if": {
                    "filter_query": "{Indicador} < 0",
                    "column_id": "Indicador",
                },
                "backgroundColor": "#FDECEA",
                "color": "#B71C1C",
                "fontWeight": "bold",
            },
        ],
    )

    return table

df_anio = pd.read_csv('rating_promedio.csv')
getIndicadorAnio(df_anio,"2024", "2025")
```

```python
import plotly.express as px

def create_line_chart(df,anio1="2020", anio2="2025"):
  
  fig = px.line(
      df,
      x='Date',
      y='Rating',
      color='Language',
      title='Tendencia de Ratings de Lenguajes de Programación (2020-2025)',
      color_discrete_sequence=px.colors.qualitative.Light24,
      markers=True
  )

  fig.update_layout(
      legend=dict(
          orientation="h",    
          yanchor="top",
          y=-0.25,
          xanchor="center",
          x=0.5
      ),
      margin=dict(b=100)       # más espacio abajo para la leyenda
  )

  return fig
```

```python
def get_monthly_winners(df, year1="2020", year2="2025"):
  df_filtered = df[(df['Year'] >= int(year1)) & (df['Year'] <= int(year2))].copy()
  df_filtered['Month'] = df_filtered["Date"].dt.strftime("%B")
  idx = df_filtered.groupby(["Year", "Month"])["Rating"].idxmax().reset_index(drop=True)
  winners = df_filtered.loc[idx, ["Year", "Month", "Language"]].copy()
  counts = (
        winners.groupby(["Language"])
        .size()
        .reset_index(name="Top1_Count").sort_values(by="Top1_Count", ascending=False)
    )
  fig = px.bar(
      counts,
      x='Language',
      y='Top1_Count',
      title=f'Número de Veces que Cada Lenguaje Fue el Mejor Calificado ({year1}-{year2})',
      color='Language',
      color_discrete_sequence=px.colors.qualitative.Pastel
  )
  return fig
get_monthly_winners(df, year1="2020", year2="2025")
```
